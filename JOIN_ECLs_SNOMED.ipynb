### Cargamos el CORE subset, actualizado con los conceptId de la versión Argentina Snapshot de SNOMED-CT, liberada por el NRC (National Release Center) Mayo 2024:

```python
# Ruta del archivo en Google Drive:
file_path = '/content/drive/MyDrive/Colab Notebooks/AMS-scripts/SNOMED-CT/CORE-MASTER-CSV-JUN2024.csv'

# Leemos el archivo en el DataFrame, especificando que las columnas numéricas se lean como texto
dtype = {
  'conceptId': str,
  }
# Leemos el archivo en un DataFrame
CORE_MASTER = pd.read_csv(file_path, delimiter=',',dtype=dtype)
```

```python
CORE_MASTER.info()
```



---



### Cargamos el archivo dummy.txt, que contiene conceptos de diagnósticos para una prestación determinada en desarrollo:

```python
# Ruta del archivo en Google Drive:
file_path5 = '/content/drive/MyDrive/Colab Notebooks/AMS-scripts/SNOMED-CT/dummy.txt'

dtype = {
    'conceptId': str,
}

# Leer el archivo en un DataFrame
dummy = pd.read_csv(file_path5, delimiter='\t', dtype=dtype)
```

```python
dummy.head(20)
```

```python
# eliminamos antes del join las columnas no deseadas:
dummy = dummy.drop(columns=['Preferred Term'])
dummy.head(20)
```

```python
dummy.info()
```



---



###Hacemos un join entre CORE_MASTER y dummy.
* Obtenemos como resultado el archivo dummy_join con los conceptId vinculados a la información del CORE:

```python
dummy_join = pd.merge(CORE_MASTER, dummy, on='conceptId', how='outer')
dummy_join.head(20)
```

###Guardamos dummy_join en el Drive:

```python
# Ruta para guardar el archivo en Google Drive
output_file_path2 = '/content/drive/MyDrive/Colab Notebooks/AMS-scripts/SNOMED-CT/dummy_join.txt'

# Guardamos el DataFrame como un archivo de texto
dummy_join.to_csv(output_file_path2, index=False, encoding='latin1', sep='\t')

print(f'Archivo guardado en: {output_file_path2}')
```
