###Archivo Descripciones Snapshot SNOMED-CT NRC ARG Mayo 2024:

*
https://drive.google.com/file/d/168IiWk-3XOe17OLrczWZb6g0qe8YQcbt/view?usp=sharing

```python
# Ruta del archivo en Google Drive:
file_path = '/content/drive/MyDrive/Colab Notebooks/AMS-scripts/SNOMED-CT/sct2_Description_Snapshot_ArgentinaEdition_20240531.txt'
'''
# Leemos el archivo en el DataFrame, especificando que las columnas numéricas se lean como texto
dtype = {
    'id': str,
    'effectiveTime': str,
    'active': str,
    'moduleId': str,
    'conceptId': str,
    'languageCode': str,
    'typeId': str,
    'term': str,
    'caseSignificanceId': str
}
'''
dtype = {
    'id': str,
    'conceptId': str,
  }
# Leer el archivo en un DataFrame
descripciones_snomed = pd.read_csv(file_path, delimiter='\t',dtype=dtype)

```

```python
# Vista previa:
descripciones_snomed.head()
```

```python
#Info:
descripciones_snomed.info()
```

```python
# Filtramos el DataFrame y eliminamos las filas con las condiciones dadas:
# conceptos inactivos (active == 0)
# Mantenemos solo las filas donde 'active' es 1
descripciones_filtradas_snomed = descripciones_snomed[descripciones_snomed['active'] != 0]

descripciones_filtradas_snomed.info()
```

```python
# Verificamos el resultado mostrando las primeras filas
descripciones_filtradas_snomed.head()
```

```python
# Filtramos el DataFrame eliminando las filas con las condiciones dadas:
# Eliminamos los conceptos en inglés (languageCode == en)
# Mantenemos solo las filas donde languageCode == es (español)
descripciones_filtradas_snomed = descripciones_filtradas_snomed[descripciones_filtradas_snomed['languageCode'] != 'en']

descripciones_filtradas_snomed.info()
```

```python
# Verificamos el resultado mostrando las primeras filas
descripciones_filtradas_snomed.head()
```

```python
# Filtramos el DataFrame eliminando las filas con las condiciones dadas:
# Conceptos sinónimos (typeId == 900000000000003001 )
# mantenemos solo las filas en donde typeId == 900000000000003001 (FSN)
descripciones_filtradas_snomed = descripciones_filtradas_snomed[descripciones_filtradas_snomed['typeId'] != 900000000000013009]

descripciones_filtradas_snomed.info()
```

```python
# Verificamos el resultado mostrando las primeras filas
descripciones_filtradas_snomed.head()
```



---



###Eliminamos las columnas que no necesitamos:
* effectiveTime       
* active              
* moduleId            
* languageCode        
* typeId          
* caseSignificanceId     

###Nos quedaría el dataframe con las siguientes columnas:
* id                  
* conceptId            
* term    

```python
# método drop para eliminar las columnas no deseadas
descripciones_filtradas_snomed = descripciones_filtradas_snomed.drop(columns=['effectiveTime', 'active', 'moduleId', 'languageCode', 'typeId', 'caseSignificanceId'])

# Verificamos resultados:
descripciones_filtradas_snomed.head()
```

```python
descripciones_filtradas_snomed.info()
```

```python
# guardamos el dataframe en Google Drive
output_file_path2 = '/content/drive/MyDrive/Colab Notebooks/AMS-scripts/SNOMED-CT/descripciones_filtradas_snomed.csv'

# Lo guardamos como un archivo .csv
descripciones_filtradas_snomed.to_csv(output_file_path2, index=False,   encoding='utf-8')

print(f'Archivo guardado en: {output_file_path2}')
```

### Cargamos el archivo de Mapeos de CIE-10:

* der2_iisssccRefset_ExtendedMapSnapshot_ArgentinaEdition_20240531.txt

```python
# Ruta del archivo en Google Drive:
file_path2 = '/content/drive/MyDrive/Colab Notebooks/AMS-scripts/SNOMED-CT/der2_iisssccRefset_ExtendedMapSnapshot_ArgentinaEdition_20240531.txt'

dtype = {
    'id': str,
    'referencedComponentId': str,
  }

# Leer el archivo en un DataFrame
cie10_refset_extendedMap = pd.read_csv(file_path2, delimiter='\t', dtype=dtype)
```

```python
cie10_refset_extendedMap.head()
```

```python
cie10_refset_extendedMap.info()
```

```python
# Filtramos el DataFrame y eliminamos las filas con las condiciones dadas:
# conceptos inactivos (active == 0)
# Mantenemos solo las filas donde 'active' es 1
cie10_refset_extendedMap = cie10_refset_extendedMap[cie10_refset_extendedMap['active'] != 0]

cie10_refset_extendedMap.info()
```

```python
cie10_refset_extendedMap.head()
```

```python
# Filtramos el DataFrame y eliminamos las filas con las condiciones dadas:
# conceptos inactivos (active == 0)
# Mantenemos solo las filas donde 'active' es 1
cie10_refset_extendedMap = cie10_refset_extendedMap[cie10_refset_extendedMap['mapGroup'] == 1]
cie10_refset_extendedMap = cie10_refset_extendedMap[cie10_refset_extendedMap['mapPriority'] == 1]
cie10_refset_extendedMap.info()
```

```python
cie10_refset_extendedMap.head()
```

###Renombramos la columna **referencedComponentId** como **conceptId** utilizando el método rename de pandas:

```python
# Renombramos la columna 'referencedComponentId' a 'conceptId'
cie10_refset_extendedMap.rename(columns={'referencedComponentId': 'conceptId'}, inplace=True)

# Verificamos el resultado mostrando las primeras filas
cie10_refset_extendedMap.info()
```



---



###Eliminamos las columnas que no necesitamos:
* id
* effectiveTime
* active
* moduleId
* refsetId
* mapGroup
* mapPriority
* correlationId
* mapCategoryId    

###Nos quedaría el dataframe con las siguientes columnas:
*	conceptId
*	mapRule
*	mapAdvice
* mapTarget


```python
# método drop para eliminar las columnas no deseadas
cie10_refset_extendedMap = cie10_refset_extendedMap.drop(columns=['id', 'effectiveTime', 'active', 'moduleId', 'refsetId','mapGroup', 'mapPriority', 'correlationId', 'mapCategoryId'])
```

```python
# Verificamos resultados:
cie10_refset_extendedMap.head(15)
```

```python
# guardamos el dataframe en Google Drive
output_file_path10 = '/content/drive/MyDrive/Colab Notebooks/AMS-scripts/SNOMED-CT/cie10_refset_extendedMap.csv'

# Lo guardamos como un archivo .csv
cie10_refset_extendedMap.to_csv(output_file_path10, index=False,   encoding='utf-8')

print(f'Archivo guardado en: {output_file_path10}')
```



---




### Hacemos un innerjoin entre **descripciones_filtradas_snomed** y **cie10_refset_extendedMap**: **snomed_cie10_inner_join**


```python
snomed_cie10_inner_join = pd.merge(descripciones_filtradas_snomed, cie10_refset_extendedMap, on='conceptId', how='inner')
```

```python
snomed_cie10_inner_join.head()
```

```python
snomed_cie10_inner_join.head()
```

```python
# Ruta para guardar el archivo en Google Drive
output_file_path = '/content/drive/MyDrive/Colab Notebooks/AMS-scripts/SNOMED-CT/snomed_cie10_inner_join.csv'

# Guardamos el DataFrame como un archivo .csv
snomed_cie10_inner_join.to_csv(output_file_path, index=False,  encoding='latin1')

print(f'Archivo guardado en: {output_file_path}')
```
